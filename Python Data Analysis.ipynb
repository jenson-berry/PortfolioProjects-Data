```python
#Import required libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
plt.style.use('ggplot')
```

```python
#Read the data
df = pd.read_csv('movies.csv')
```

# 2. EDA & Cleaning

```python
#Exploratory Data Analysis

df.head()
```

```python
df.info()
```

## 2.1 Data Cleaning

### 2.1.1 Missing Data

```python
for column in df.columns:
    pct_missing = np.mean(df[column].isnull())
    print(f"{column} - {pct_missing}%")
```

```python
df = df.dropna()
```

### 2.1.2 Data Types

```python
df.dtypes
```

```python
#Change budget and gross to int

df[['budget','gross']] = df[['budget','gross']].astype('int64')
```

### 2.1.3 Correct Year (Release Date)

```python
#Create new column for correct year

df['correctYear'] = df['released'].str.extract(pat ='([0-9]{4})').astype(int)
```

```python
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

### 2.1.4 Duplicate Rows / Values

```python
df.drop_duplicates() #not applying to DF
```

# 3. Data Correlation

## 3.1 Budget vs Gross

### 3.1.1 Scatterplot

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget ')
plt.ylabel('Gross Earnings')
```

```python
sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color": "green"})
```

### 3.1.2 Correlation

```python
# Correlation
df.corr()
```

```python
corr_matrix = df.corr()
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Variables')
plt.xlabel('Movie Variables ')
plt.ylabel('Movie Variables')
plt.show()
```

## 3.2 Company vs. Gross

```python
df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes

df_numerised
```

```python
corr_matrix = df.corr()

sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Variables')
plt.xlabel('Movie Variables ')
plt.ylabel('Movie Variables')
plt.show()
```

### 3.1.3 Other Correlations

```python
correlation_mat = df_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
```
